To implement linear regression for predecting continous target variable and evaluate the model perfromance using appropriate metrics


```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


```

```python
housing = fetch_california_housing()

df=pd.DataFrame(housing.data, columns=housing.feature_names)
df['target']=housing.target
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
x = df.drop('target',axis = 1)
y = df['target']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
y_test.shape
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mae = mean_squared_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,y_pred)
```

```python
print("\nModel Evaluation Metrics:")
print("------------------------------------------------------")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (R2): {r2}")
```

```python
plt.scatter(x_test['MedInc'],y_test,color='yellow')
plt.scatter(x_test['MedInc'],y_pred,color='red')
plt.xlabel('MedInc')
plt.ylabel('Target')
plt.title('Linear Regression')
plt.show()
```
